## 2. 処理：データ型を変えずに穴埋め

```python
# 01_pattern_list_to_vertical.py
dfm = pd.DataFrame(
    [
    {"Pos":"QB","regular":12,"Reserve":[9,8]},
    {"Pos":"CB","regular":28,"Reserve":[37,23,38,20]},
    {"Pos":"WR","regular":19,"Reserve":[83,82,81,17,18]},
    {"Pos":"T","regular":69,"Reserve":[75,78]},
    {"Pos":"S","regular":36,"Reserve":[35,21,29,27]},
    {"Pos":"RB","regular":22,"Reserve":[33,88,30]}
    ]
)

to_vertical_columns = "Reserve"
outname = 'pattern_list_to_vertical.csv'

# DataFrameの入れ物（dfm_vertical）を作成
dfm_vertical = pd.DataFrame(columns=dfm.columns)

# 縦持ち展開したい要素それぞれを抽出
for i in range(len(dfm)):
    # 要素内のリストの各要素を抽出
    for j in range(len(dfm.loc[i, to_vertical_columns])):
        # リストを1要素に変換した行を作成
        dfm_tmp = dfm.loc[i].copy()
        dfm_tmp[to_vertical_columns] = [dfm_tmp.loc[to_vertical_columns][j]]
        # 縦持ち版データフレームに新たに作成した行を追加
        dfm_vertical = dfm_vertical.append(dfm_tmp)
        
        # indexを振り直す
        dfm_vertical = dfm_vertical.reset_index(drop=True)


dfm_vertical
```
